#### Problem 1: Linked Lists

Using the above examples as a guide, create your own interpretation of the a Linked List class. You can not use the code above exactly, but again it can be used as a guide. This problem requires you to think about how a linked list works and create one using your own logic.

*Remember* A Linked List is a list of Nodes that point to the next node in the chain. The first Node starts out as Empty(None) and each node after points to the next.

Your Linked List should have a traverse method and have the ability to add a new node

```python
class Song():
    def __init__(self, value):
        self.value = value
        self.next = None
        
    def __str__(self):
        return f"Song: {self.value}"
    
    
class PlaylistOrder():
    def __init__(self):
        self.head = None
        
    # add song to beginning of playlist
    def add_to_beginning(self, next_song):
        new_song = Song(next_song)
        new_song.next = self.head
        self.head = new_song
        
    # add song after a certain song
    def play_after(self, prev_song, next_song):
        
        # is there a prvious song in the playlist:
        if prev_song is None:
            print("There is no song before this one. ")
            return
        
        # if there's a song before, insert the one you want after it
        new_song = Song(next_song)
        
        # song ends, play the next one
        new_song.next = prev_song.next
        
        # song playing is the one after the one that just played
        prev_song.next = new_song
        
    # append with a song
    def add_to_end(self, next_song):
        new_song = Song(next_song)
        
        # is playlist empty? if yes:
        if self.head is None:
            
            # add song to playlist
            self.head = new_song
            
        # if there's songs already, put it at the end
        last_song = self.head
        
        # keep going to find where to put it
        while last_song.next:
            last_song = last_song.next
            
        # make the last song the one you want to add to the end
        last_song.next = new_song
        
    # show whole playlist
    def traverse(self):
        temp = self.head
        
        # keep going to find the None value
        while temp:
            print(temp)
            temp = temp.next
            
            
playlist = PlaylistOrder()

def main():
    while (response:=input("What do you want to do? \nAdd to beginning (add), insert one inbetween two (insert), add song to end (end), or show playlist (show)? ")):
        if response.lower() == "add":
            input("What song would you like to add? ")
            playlist.add_to_beginning()
            playlist.traverse()
            
        if response.lower() == "insert":
            input("What song would you like to add? ")
            playlist.play_after()
            playlist.traverse()
            
        if response.lower() == "end":
            input("What song would you like to add? ")
            playlist.add_to_end()
            playlist.traverse()
            
        if response.lower() == "show":
            playlist.traverse()
            
        else:
            print("Not a valid command. ")
            playlist.traverse()
            break

main()
```

#### Problem 2: Binary Search Tree

Using the above examples as a guide, create your own interpretation of the a Binary Search Tree class. You can not use the code above exactly, but again it can be used as a guide. This problem requires you to think about how a Binary Search Tree works and create one using your own logic.

*Remember* Binary Search Trees start with a head node and each node to the left of that will be smaller, each node to the right of it will be greater. The far left node should be the lowest number(if one exists) that is available. The far right node (if one exists) should be the greatest number

```python
class BST:
    def __init__(self, value):
        self.left_child = None
        self.right_child = None
        self.data = value

# adding values
def insert_number(root, node):
    if root is None:
        root = node
    else:
        if root.data > node.data:
            if root.left_child is None:
                root.left_child = node
            else:
                insert_number(root.left_child, node)
        else:
            if root.right_child is None:
                root.right_child = node
            else:
                insert_number(root.right_child, node)

# printing the list of numbers
def new_list(root):
    if not root:
        return
    new_list(root.left_child)
    print(root.data)
    new_list(root.right_child)

start = BST(2)
insert_number(start, BST(8))
insert_number(start, BST(1))
insert_number(start, BST(4))
insert_number(start, BST(10))
insert_number(start, BST(124))
insert_number(start, BST(3))

print("Numbers, in order: ")
new_list(start)
```
